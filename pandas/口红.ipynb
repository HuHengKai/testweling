```python
data=pd.read_csv('lipstick.csv')
```

```python
data['deal']=data['deal'].str.strip( '人付款' )
data['deal']=data['deal'].astype('float')
```

```python
data['price']=data['price'].apply(lambda x: x[1:]).astype('float')
```

```python
data['amount'] = data.price * data.deal
```

```python
data['location_1']=data['location'].apply(lambda x: x[0:3])
```

```python
data.drop(["image_src"],axis=1)
```

```python
plt.figure(figsize=(8,6))
figure_shop=data.groupby("shop").deal.sum()
```

```python
figure_shop=figure_shop.sort_values(ascending=False)[:30]
figure_shop.plot(kind="bar")
```
