```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('数据 (1).csv',encoding="GBK")
#遇到有中文不能读取时候，使用encoding指定编码格式
```

```python
#获取某一月得数据
data['month']=pd.to_datetime(data['支付时间'])
month3_data=data[data['month'].dt.month==3]
month3_data
```

```python
amount=month3_data['商品'].value_counts()
names=amount[0:5]
```

```python
plt.plot(names)
```
