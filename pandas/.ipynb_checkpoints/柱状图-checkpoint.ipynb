```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('数据 (1).csv',encoding="GBK")
#遇到有中文不能读取时候，使用encoding指定编码格式
```

```python
#获取某一月得数据
data['month']=pd.to_datetime(data['支付时间'])
month3_data=data[data['month'].dt.month==3]
month3_data
```

```python
amount=month3_data['商品'].value_counts()

names=amount.index[0:10]
names
```

```python
account=amount.values[0:10]
account2=[50,50,60,60,60,70,80,90,10,28]
account2
```

```python
plt.figure(figsize=(20,8))
plt.bar(names,account,width=0.3,color="goldenrod")

plt.bar(names,account2,width=0.3,color="red")
plt.xlabel('x-names',fontsize=14)
plt.ylabel('y-account',fontsize=14)
plt.show()
```

```python
place=list(set(data['地点']))
#获取所有地点
place
```

```python
months=[i for i in range(1,5)]
month_name=['一月', '二月', '三月', '四月']
data['months']=pd.to_datetime(data['支付时间'])
data.head()
```

```python
for pla in place:
    place_data=data[data['地点']==pla]
    place_money=list()
    for mon in months:
        month_data=place_data[place_data['months'].dt.month==mon]
        money=month_data['金额'].sum()
        place_money.append(money)
    print(place_money)
    plt.plot(months, place_money, label=pla+'地区')
```
