```python
#pd.concat()只是单纯得把两个表拼接再一起，参数aix用于指定行还是列。默认为0,为行连接
#axis=1为列连接。效果相当于append
pd.concat([stocks1,stocks2,stocks3],axis=0,ignore_index=True)
#类似于关系型数据库的连接方式，可以根据一个或多个键将不同的DatFrame连接起来。该函数的典型应用场景是，针对同一个主键存在两张不同字段的表，根据主键整合到一张表里面
#理解为列连接
```
